```python


# include only California wines.

california_wines = [wine for wine in wine_data[1:] if wine[5] == "California"]

# Extract the points column from those wines.

points_list = [int(wine[3]) for wine in california_wines]

#find average

average_points = sum(points_list) / len(points_list)

print("Average rating for California wines:", average_points)
```

```python

- What's the most expensive wine from Spain ?'

# include only wine from spain 
spain_wines = [wine for wine in wine_data[1:] if wine[1] == "spain"]

#Extract the price column from those wines.

price_list = [float(wine[4]) for wine in spain_wines if wine[4] != '']

# compare price

max_price = max(price_list)

most_expensive_wines = [wine for wine in spain_wines if wine[4] != '' and float(wine[4]) == max_price]

print("Most expensive wine(s) from Spain:")
```

```python
- Where's the worse wine in the US from ?

# include only wine from the US 

US_wines = [wine for wine in wine_data[1:] if wine[1] == "US"]

# Extract the points column from those wines.

points_list = [int(wine[3]) for wine in US_wines]


# find minimum
min_points = min(points_list)

worst_wine = [wine for wine in US_wines if int(wine[3]) == min_points]

print("Worst wine from US:")
for wine in worst_wine:
    print(wine[5], "-", wine[6], "with", wine[3], "points")
```

```python

- Which country between spain/us/france has highest average quality points ?

# include only wine from spain, us france
selected_wines = [wine for wine in wine_data[1:] if wine[1] in ["Spain", "US", "France"]]

# Extract the points column from each of these wines.
country_points = {}

for wine in selected_wines:
    country = wine[1]
    points = int(wine[3])
    
    if country not in country_points:
        country_points[country] = {"sum": 0, "count": 0}
    
    country_points[country]["sum"] += points
    country_points[country]["count"] += 1



# find average
country_avg = {country: data["sum"]/data["count"] for country, data in country_points.items()}

#find maximum
best_country = max(country_avg, key=country_avg.get)
print("Country with highest average quality points:", best_country)
```

```python
def filter_by_country(all_data, which_country):
  filtered_items = []

  for item in all_data:
    if item[1] == which_country:
      filtered_items.append(item)

  return filtered_items
```

```python
wines_us = filter_by_country(wine_data[1:], "US")
wines_sp = filter_by_country(wine_data[1:], "Spain")

display(len(wines_us))
display(len(wines_sp))
```
