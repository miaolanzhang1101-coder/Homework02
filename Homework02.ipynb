### Example Filter

Separate all _A+_ wines from _Oregon_ (assuming _A+_ means more than $95$ points):

```python
# Initialize variable based on what we want the result to be
# In this case, a list of wines
good_or = []

# skip header
for wine in wine_data[1:]:
  if wine[5] == "Oregon" and float(wine[3]) > 95:
    good_or.append(wine)

print(len(good_or))
print(good_or)
```
