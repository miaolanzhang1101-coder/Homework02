```python
# TODO: separate prices into 3 lists, one for each country
# TODO: calculate the average (sum / length)

wines_sp = []
wines_us = []

for wine in wine_data[1:]:
  if wine[1] == "Spain":
    wines_sp.append(wine)
  if wine[1] == "US":
    wines_us.append(wine)

display(len(wines_sp))

sum_sp = 0
for wine in wines_sp:
  sum_sp = sum_sp + float(wine[4])

sum_sp / len(wines_sp)
```

```python
def filter_by_country(all_data, which_country):
  filtered_items = []

  for item in all_data:
    if item[1] == which_country:
      filtered_items.append(item)

  return filtered_items
```

```python
wines_us = filter_by_country(wine_data[1:], "US")
wines_sp = filter_by_country(wine_data[1:], "Spain")

display(len(wines_us))
display(len(wines_sp))
```

```python
sum_sp = 0
cnt_sp = 0

for wine in wine_data[1:]:
  if wine[1] == "Spain":
    sum_sp = sum_sp + float(wine[4])
    cnt_sp = cnt_sp + 1

display(sum_sp)
display(sum_sp / cnt_sp)
```
